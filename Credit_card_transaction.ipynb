### Data Manipulation and feature engineering

```python
df['Discounted_Amount'] = np.where(df['Amount'] > 500, round(df['Amount'] - (df['Amount'] * 0.5)), df['Amount'])
```

```python
df[['Transaction_ID', 'Amount', 'Discounted_Amount']].head()
```

```python
df.loc[df['Amount'] < 500, ['Transaction_ID', 'Amount', 'Discounted_Amount']]
```

```python
df['Transaction_Amt_Category'] = np.where(df['Amount'] > 500, 'High', np.where(df['Amount'] > 100, 'Medium', 'Low'))
```

```python
df.loc[(df['Amount'] <100), ['Transaction_ID', 'Amount', 'Transaction_Amt_Category']]
```

### Fraud Detection Inidicators

```python
df.head()
```

```python
result = df.groupby(['Customer_ID', 'Transaction_Date']).agg({'Transaction_ID': 'count'})
result.columns = ['Transaction_Count']
result['Transaction_Count'].unique()
```

As there is no transaction more than 1 or not equal to or greater than 10. So no customer flagged as fraud

```python
df.info()
```

Only Counting the location within same date will conclude the transaction happened by same customer on same date but on different locations. (Each location is associated with the only unique transaction happened)

```python
result = df.groupby(['Customer_ID', 'Transaction_Date']).agg({'Location': 'count'})
print(result['Location'].unique())
result
```

The max transaction happend on the same date by same customer is 2

```python
result.columns = ['Location_count']
result[result['Location_count'] > 1] 
```

```python
#Find transactions where Amount > $5000 and Transaction_Type is Online (flag as high-risk).

df.loc[(df['Transaction_Type'] == 'Online') & (df['Amount'] > 5000)]
```

Hence no customer found. No customer flagged fraud with high-risk
